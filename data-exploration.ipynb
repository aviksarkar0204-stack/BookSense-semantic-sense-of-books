```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(f"{path}/books.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
x=plt.axes()
sns.heatmap(df.isnull().transpose(), cbar=False, ax=x)
plt.xlabel("Column")
plt.ylabel("Missing Values")
```

```python
df["missing_description"] = np.where(df["description"].isna(), 1, 0)
df["age_of_book"] = 2025 - df["published_year"]
```

```python
columns_of_interest = ["num_pages","age_of_book","missing_description","average_rating"]

correlation_matrix = df[columns_of_interest].corr(method="spearman")

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",cbar_kws={"label": "Spearman's Rank Correlation"})
heatmap.set_title("Correlation Matrix")
plt.show()
```

```python
df[(df["description"].isna()) |
   (df["num_pages"].isna()) |
   (df["average_rating"].isna()) |
   (df["published_year"].isna())
]
```

```python
df_missings = df[~(df["description"].isna()) &
   ~(df["num_pages"].isna()) &
   ~(df["average_rating"].isna()) &
   ~(df["published_year"].isna())
]
```

```python
df_missings_25_words = df_missings[df_missings["Words_in_description"] >= 25]
```

```python
df_missings_25_words
```

```python
df_missings_25_words["title_and_subtitle"] = (np.where(df_missings_25_words["subtitle"].isna(), df_missings_25_words["title"], df_missings_25_words[["title","subtitle"]].astype(str).agg(": ".join, axis=1))
                                              )
```

```python
df_missings_25_words
```

```python
df_missings_25_words["tagged_description"] = df_missings_25_words[["isbn13","description"]].astype(str).agg(" ".join, axis=1)
```
